```python
#Beginning of Problem 3 Solution


def integrate(lowerbound,upperbound,N):
    I=0
    step = (upperbound-lowerbound)/(N)
    for i in range(1,N):
        x_i=lowerbound+i*step
        f=m.sin(x_i)
        I=I+f*step   
    return(I)

for j in range(1,6):
    print("Integration result for", 10**j, "rectangles", integrate(0,m.pi/2,10**j))


for k in range(1,6):
    print("Absolute error for",10**k,"rectangles:",(1-integrate(0,m.pi/2,10**k)))

for l in range(1,5):
    print("Ratio of Absolute error for",10**l,"rectangles to", 10**(l+1), "rectangles", ((1-integrate(0,m.pi/2,10**l)))/((1-integrate(0,m.pi/2,10**(l+1)))))

#As seen in the ratios of the absolute errors, the graph that yields absolute error as a function of number of rectangles
#has a constant slope of around 10. This suggests that the integration method is a first-order method.



```
