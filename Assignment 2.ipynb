```python
#Name: Evan Manley , UNI: ejm2237
#Problem 1: Implementing a Higher Order Integrator
import math as m
import numpy as np

#Thanks for the tip of making the function general. I believe I have made this one general by allowing the integrand
#to be passed into the function.

def integrate(integrand,lowerbound,upperbound,N):
    I=0
    step = (upperbound-lowerbound)/(N)
    for i in range(0,N):
        x_i=lowerbound+i*step #The ith point of evaluation
        x_i1=lowerbound+(i+1)*step #The (i+1)th point of evaluation
        f_left=integrand(x_i) 
        f_mid=integrand((x_i+x_i1)/2)
        f_right=integrand(x_i1)
        I=I+step*((1/6)*f_left+(2/3)*f_mid+(1/6)*f_right)   
    return(I)

for j in range(1,6):
    print("Integration result for", 10**j, "steps", integrate(m.sin,0,m.pi/2,10**j))
    
for k in range(1,6):
    print("Fractional error for",10**k,"steps:",(1-integrate(m.sin,0,m.pi/2,10**k)))

for l in range(1,5):
    print("Ratio of fractional error for",10**l,"steps to", 10**(l+1), "steps", ((1-integrate(m.sin,0,m.pi/2,10**l)))/((1-integrate(m.sin,0,m.pi/2,10**(l+1)))))

#This method is much more accurate than the first order integration method in the last problem set.
#The ratio of fractional errors being around 10000 for a increase in the number of steps by a factor of 10 indicates
#a fourth order integration method.
    
```

```python
#Problem 2: A Limb-darkening Flux Function

def delta(p,r,z):
    if r>=(z+p) or r<=(z-p):
        delta_function=0
        return(delta_function)
    elif (r+z)<=p:
        delta_function=1
        return(delta_function)
    else:
        delta_function=(1/m.pi)*m.acos((z**2-p**2+r**2)/(2*z*r))
        return(delta_function)
    r=r+.01


#ldf stands for limb darkening function    
def numerator_integral(ldf, lowerbound, upperbound, N, p, z):
    I=0
    step = (upperbound-lowerbound)/(N)
    for i in range(1,N):
        x_i=lowerbound+i*step
        x_i1=lowerbound+(i+1)*step
        f_left=ldf*(1-delta(p,x_i,z))*2*x_i
        f_mid=ldf*(1-delta(p,(x_i+x_i1)/2,z))*(x_i+x_i1)
        f_right=ldf*(1-delta(p,x_i1,z))*2*x_i1
        I=I+step*((1/6)*f_left+(2/3)*f_mid+(1/6)*f_right)   
    
    return I


def denominator_integral(ldf, lowerbound, upperbound, N,p,z):
    I=0
    step = (upperbound-lowerbound)/(N)
    for i in range(1,N):
        x_i=lowerbound+i*step
        x_i1=lowerbound+(i+1)*step
        f_left=ldf*(2*(x_i))
        f_mid=ldf*(x_i+x_i1)
        f_right=ldf*(2*x_i1)
        I=I+step*((1/6)*f_left+(2/3)*f_mid+(1/6)*f_right)   
    return(I)

def flux(p, z, ldf, lowerbound, upperbound, N):
    return numerator_integral(ldf, lowerbound,upperbound,N,p,z) / denominator_integral(ldf, lowerbound, upperbound, N,p,z)

for j in range(1,6):
    print(flux(0.2, 0.9, 1, 0, 1, 10**j))
    
#The analytic formula in the first problem set yields a flux value of .968417005 for p=.2 and z=.9.
for k in range(1,6):
    print("Fractional error for",10**k,"steps:",(.968417005-flux(0.2, 0.9, 1, 0, 1, 10**k))/(.968417005))

for l in range(1,5):
    print("Ratio of fractional error for",10**l,"steps to", 10**(l+1), "steps", ((.968417005-flux(0.2, 0.9, 1, 0, 1, 10**l))/(.968417005))/((.968417005-flux(0.2, 0.9, 1, 0, 1, 10**(l+1)))/(.968417005)))

#My suitemate Nixon McKenzie who is a Junior in the Data Science major offered to help optimize code that I had already
#written for evaluating flux. He suggested that I change the 
#print(numerator_integral() / denominator_integral()) 
#command at the end of my original code to a separate function so that the flux function can be easily called should it be
#needed later in the code.
```

```python
#Problem 3

import random
 

def MC_Flux(N,p,z):
    N1=0
    N2=0
    for i in range(0,N):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if (x**2+y**2)<1:
            N1=N1+1
        else:
            N1=N1
        if (x-z)**2+y**2<p**2:
            N2=N2+1
        else:
            N2=N2
    return((N1-N2)/N1)

        

for j in range(1,6):
    print(MC_Flux(10**j,.2,.9))
    
for k in range(1,6):
    print("Fractional error for",10**k,"points:",(.968417005-MC_Flux(10**k,.2,.9))/(.968417005))

for l in range(1,5):
    print("Ratio of fractional error for",10**l,"points to", 10**(l+1), "points", ((.968417005-MC_Flux(10**l,.2,.9))/(.968417005))/((.968417005-MC_Flux(10**(l+1),.2,.9))/(.968417005)))

#The order of error for the Monte Carlo integration is not definite. This is evident by looking at the non-steady
#behavior of the ratio of fractional error.
```
