```python
#Problem 1a
import pickle
import matplotlib.pyplot as plt
import math as m
import numpy as np
f97 = open("KOI97_01.p", "rb") #open for binary read
data97 = pickle.load(f97)
time, flux, flux_err = data97

#Problem 1b
plt.plot(time, flux)
plt.title('Kepler Light Curve', fontsize=15)
plt.xlabel('Time (UTC) - 2454833', fontsize=10)
plt.ylabel('Flux [Counts/Second]', fontsize=10)
plt.errorbar(time, flux, flux_err, ecolor='blue')
plt.show
```

```python
#Problem 2a

plt.plot((time[1300:1900]-261.3)*10, flux[1300:1900])
                                            #I played around with the arrays and chose to slice them at the first eclipse in
                                            #the data between 1300 and 1900.
plt.title('Kepler Light Curve', fontsize=15)
plt.xlabel('Time (UTC) - 2454833', fontsize=10)
plt.ylabel('Flux [Counts/Second]', fontsize=10)
plt.errorbar((time[1300:1900]-261.3)*10, flux[1300:1900], flux_err[1300:1900], ecolor='blue')
plt.show()

z_arr=np.array([])  #Here, I created an array for the z values
for i in range(0,600):
     z_arr = np.append(z_arr, -1.2+i*.004)
        
def flux_func(z,p):   #This is very similar to my flux function from the first problem set. However, I changed it to
                    #Generate an array so it'd be easier to manipulate and plot.
    flux_arr=np.array([])
    while z<1.2:
        if abs(z)>(1+p):
            function=1-0
            flux_arr = np.append(flux_arr, function)
        elif abs(1-p)<abs(z)<=(1+p):
            K1=m.acos((1-p**2+z**2)/(2*abs(z)))
            K0=m.acos((p**2+z**2-1)/(2*p*abs(z)))
            function=1-((1/np.pi)*((p**2)*K0+K1-m.sqrt((4*z**2-(1+z**2-p**2)**2)/(4))))
            flux_arr = np.append(flux_arr, function)           
        elif abs(z)<=(1-p):
            function=1-p**2
            flux_arr = np.append(flux_arr, function)
        elif abs(z)<=(p-1):
            function=1-1
            flux_arr = np.append(flux_arr, function)
            
        
       
        z=z+.004 #I made the step size .004 between -1.2 and 1.2
                    #to match the size of the sliced flux array and z value array.
    return(flux_arr)

plt.title('Theoretical Kepler Light Curve', fontsize=15)
plt.xlabel('Time (UTC) - 2454833', fontsize=10)
plt.ylabel('Flux [Counts/Second]', fontsize=10)
plt.xlim([-2,2])
plt.plot(z_arr, 106600*flux_func(-1.2, .0866)) #I used the fact that the minimum of the theoretical
                                                #flux curve will occur at a z value
                                                # that uses function = 1-p**2 and eyeballed the data curve to estimate 
                                                #function = 105800/106600 which gave p=.0866.
plt.show()
```

```python
#Problem 3

import math as m

#I defined some helper functions to make the bisection function less messy in the end.

def L_ratio(T_eff):
    return(((T_eff)/(5780))**4)

def S_eff(a,b,T_eff):
    return(1.014+(a)*(T_eff-5780)+(b)*(T_eff-5780)**2)
    
def habitable_distance(a,b,T_eff):
    return(m.sqrt((L_ratio(T_eff))/(S_eff(a,b,T_eff))))

def Func_for_Bisec(a,b,d,T_eff):
    return(habitable_distance(a,b,T_eff)-d)

def Bisection(a,b,d,func_for_bisec,epsilon,lowerbound,upperbound): 
    midpoint=(lowerbound+upperbound)/2
    while (upperbound-lowerbound)>epsilon: #while loop will narrow down bounds to a desired width of epsilon
        if func_for_bisec(a,b,d,upperbound)*func_for_bisec(a,b,d,midpoint)>=0:
            upperbound=midpoint
        else:
            lowerbound=midpoint
        midpoint=(lowerbound+upperbound)/2
    print('The Estimated Effective Temperature is:',midpoint,"K")

Bisection(8.177*10**(-5),1.706*10**(-9),.5,Func_for_Bisec,.000001,3000,5000) #The temperature has to be cooler than that
                                                                        #of the sun, so I chose bounds 3000 and 5000.

#Seems like the around the temperature I would expect! 
```
