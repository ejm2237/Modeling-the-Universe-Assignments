```python
#Problem 1

import math as m
import matplotlib.pyplot as plt
import numpy as np
import scipy.constants as sp

class Eq_State:
    
    def __init__(self, rho_0, c, n):
        self.rho_0 = rho_0
        self.c = c
        self.n = n
    
    def value(self, P):
        return(self.rho_0 + (self.c)*P**(self.n))
    
def eq_state(rho_0, c, n, P):
    return(Eq_State(rho_0, c, n).value(P))

def dPdr(r, M, P):
    return(-((sp.G*M*eq_state(8300, .00349, .528, P))/r**2))

def dMdr(r, P):
    return (4*m.pi*eq_state(8300, .00349, .528, P)*r**2)

def RK4(M, r, P, h, rho_0, c, n):
    P_array = np.array([P])
    M_array = np.array([M])
    r_array = np.array([r])
    rho_array = np.array([eq_state(rho_0, c, n, P)])
    
    while P>0:
        k1 = dPdr(r, M, P)
        l1 = dMdr(r, P)
        k2 = dPdr(r + h/2, M + l1*h/2, P + k1*h/2)
        l2 = dMdr(r + h/2, P + k1*h/2)
        k3 = dPdr(r + h/2, M + l2*h/2, P + k2*h/2)
        l3 = dMdr(r + h/2, P + k2*h/2)
        k4 = dPdr(r + h, M + l3*h, P + k3*h)
        l4 = dMdr(r + h, P + k3*h)
        P = P + np.real((1/6)*(k1 + 2*k2 + 2*k3 + k4)*h) #was getting complex valued numbers here, for some reason so I had 
                                                #to take the real part. I investigated the complex numbers and found that 
                                                        #the imaginary component was negligible.
        M = M + np.real((1/6)*(l1 + 2*l2 + 2*l3 + l4)*h)
        rho = eq_state(rho_0, c, n, P)
        r = r + h
        P_array = np.append(P_array, P)
        M_array = np.append(M_array, M)
        rho_array = np.append(rho_array, rho)
        r_array = np.append(r_array, r)
        
    plt.plot(r_array, P_array)
    plt.title("Planetary Pressure Radius Graph")
    plt.ylabel("Pressure [Pa]")
    plt.xlabel("Radius [m]")
    plt.show()
    plt.plot(r_array, M_array)
    plt.title("Planetary Mass Radius Graph")
    plt.ylabel("Mass [kg]")
    plt.xlabel("Radius [m]")
    plt.show()
    plt.plot(r_array, rho_array)
    plt.title("Planetary Density Radius Graph")
    plt.ylabel("Density [kg/m^3]")
    plt.xlabel("Radius [m]")
    plt.show()
    
    print("The radius of the planet is approximately:", r/(6371000), "Earth Radii")
    print("The mass of the planet is approximately:", M/(5.972e+24), "Earth Masses")

RK4(0, 1, 10**12, 5000, 8300, .00349, .528)
    

        
        
        
        

    
```

```python
#Problem 2

def Max_Mass_Func(M, r, P, h, rho_0, c, n):
    
    P_array = np.array([P])
    M_array = np.array([M])
    r_array = np.array([r])
    rho_array = np.array([eq_state(rho_0, c, n, P)])
    
    while P>0:
        k1 = dPdr(r, M, P)
        l1 = dMdr(r, P)
        k2 = dPdr(r + h/2, M + l1*h/2, P + k1*h/2)
        l2 = dMdr(r + h/2, P + k1*h/2)
        k3 = dPdr(r + h/2, M + l2*h/2, P + k2*h/2)
        l3 = dMdr(r + h/2, P + k2*h/2)
        k4 = dPdr(r + h, M + l3*h, P + k3*h)
        l4 = dMdr(r + h, P + k3*h)
        P = P + np.real((1/6)*(k1 + 2*k2 + 2*k3 + k4)*h)
        M = M + np.real((1/6)*(l1 + 2*l2 + 2*l3 + l4)*h)
        rho = eq_state(rho_0, c, n, P)
        r = r + h
        P_array = np.append(P_array, P)
        M_array = np.append(M_array, M)
        rho_array = np.append(rho_array, rho)
        r_array = np.append(r_array, r)
        
    return M/(5.972e+24)

def Max_Radius_Func(M, r, P, h, rho_0, c, n):
    
    P_array = np.array([P])
    M_array = np.array([M])
    r_array = np.array([r])
    rho_array = np.array([eq_state(rho_0, c, n, P)])
    
    while P>0:
        k1 = dPdr(r, M, P)
        l1 = dMdr(r, P)
        k2 = dPdr(r + h/2, M + l1*h/2, P + k1*h/2)
        l2 = dMdr(r + h/2, P + k1*h/2)
        k3 = dPdr(r + h/2, M + l2*h/2, P + k2*h/2)
        l3 = dMdr(r + h/2, P + k2*h/2)
        k4 = dPdr(r + h, M + l3*h, P + k3*h)
        l4 = dMdr(r + h, P + k3*h)
        P = P + np.real((1/6)*(k1 + 2*k2 + 2*k3 + k4)*h)  
        M = M + np.real((1/6)*(l1 + 2*l2 + 2*l3 + l4)*h)
        rho = eq_state(rho_0, c, n, P)
        r = r + h
        P_array = np.append(P_array, P)
        M_array = np.append(M_array, M)
        rho_array = np.append(rho_array, rho)
        r_array = np.append(r_array, r)
        
    return(r/(6371000))

Max_Radius_array = np.array([])
Max_Mass_array = np.array([])

for P in range(round(9.85**11), round(10.27**14), 100000000000):
    Max_Radius_array = np.append(Max_Radius_array, Max_Radius_Func(0, 1, P, 50000, 8300, .00349, .528))
    Max_Mass_array = np.append(Max_Mass_array, Max_Mass_Func(0, 1, P, 50000, 8300, .00349, .528))
    

plt.plot(Max_Radius_array, Max_Mass_array)
plt.title("Planetary Mass Radius Relation")
plt.ylabel("Log Scaled Mass [Earth Masses]")
plt.xlabel("Log Scaled Radius [Earth Radii]")
plt.xscale('log')
plt.yscale('log')
plt.show()
    
#The mass radius relation shows that it is easier to detect high mass planets because a higher mass corresponds to a larger 
#radius, which corresponds to a greater area of the star being blocked out and therefore a larger fractional
#change in the flux from the star being observed.
```
