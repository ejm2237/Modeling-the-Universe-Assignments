```python
import numpy as np
import random
import itertools as itr
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
```

```python
X = np.genfromtxt("RRLyrae_features.txt", delimiter=',')
Y = np.genfromtxt("RRLyrae_labels.txt")
    
    
#Create a scatter plot where the positions represent the numerical values of the first two colors and the color of the point
#represents the label of the point. Orange refers to RR Lyrae and black refers to other stars.
x_array_0 = np.array([])
y_array_0 = np.array([])
x_array_1 = np.array([])
y_array_1 = np.array([])

for i in range(0, len(X)):
    if Y[i] == 0:
        x_array_0 = np.append(x_array_0, X[i,0])
    else:
        x_array_1 = np.append(x_array_1, X[i,0])
        
for i in range(0, len(X)):
    if Y[i] == 0:
        y_array_0 = np.append(y_array_0, X[i,1])
    else:
        y_array_1 = np.append(y_array_1, X[i,1])
   


plt.scatter(x_array_0, y_array_0, c='black', s=2)
plt.scatter(x_array_1, y_array_1, c='orange', s=2)
plt.show()
```

```python
#Problem 1
#split the data into a training set and testing set.
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .25)


#generate 5 folds of indices and store them in a list.
indices = KFold(n_splits = 5) 
Folds = [[], [], [], [], []]

index_list = []
for i in indices.split(X_train):
    index_list = np.append(index_list, i)

for i in range(0,5):
    for j in range(0,len(index_list[2*i+1])):
        Folds[i] = np.append(Folds[i], index_list[2*i+1][j])

#apply the indices given in Folds to X_train and Y_train values
X_folds = [[], [], [], [], []]
Y_folds = [[], [], [], [], []]
for i in range(0,len(Folds)):
    for j in range(0,len(Folds[i])):
        X_folds[i] = np.append(X_folds[i], X_train[int(Folds[i][j])])
        Y_folds[i] = np.append(Y_folds[i], Y_train[int(Folds[i][j])])
    X_folds[i] = np.split(X_folds[i], len(X_folds[i])/4)


#the goal here is to create a list whose ith entry has all values of X_train except X_folds[i]. Same thing for Y_train.      
x_train_data = [[], [], [], [], []]
y_train_data = [[], [], [], [], []]

#scan through X_folds and append X_folds[j][k] to x_train_data[i] for all k in X_folds[j] only if i != j. Same thing for Y.
for i in range(0,len(X_folds)):
    for j in range(0,len(X_folds)):
        if i == j:
            pass
        else:
            for k in X_folds[j]:
                x_train_data[i] = np.append(x_train_data[i], k)
    x_train_data[i] = np.split(x_train_data[i], len(x_train_data[i])/4)

for i in range(0,len(Y_folds)):
    for j in range(0,len(Y_folds)):
        if i == j:
            pass
        else:
            for k in Y_folds[j]:
                y_train_data[i] = np.append(y_train_data[i], k)

#iterate over a range of min_samples_leaf to optimize the hyperparameter
model_list = []
for msl in range(30,50):
    dtc = DecisionTreeClassifier(min_samples_leaf = msl)
    #Train the DTC on 4 of 5 folds and test on the 5th for each fold. Compute the mean of the test results.
    predict_list = [[], [], [], [], []]
    for i in range(0,5):
        dtc.fit(x_train_data[i], y_train_data[i])
        predict_list[i] = dtc.predict(X_folds[i])


    TP_list = [0,0,0,0,0]
    TN_list = [0,0,0,0,0]
    FP_list = [0,0,0,0,0]
    FN_list = [0,0,0,0,0]

    for i in range(0, len(predict_list)):
        for j in range(0,len(predict_list[i])):
            if predict_list[i][j] == Y_folds[i][j] and Y_folds[i][j] == 1:
                TP_list[i] = TP_list[i] + 1
            elif predict_list[i][j] == Y_folds[i][j] and Y_folds[i][j] == 0:
                TN_list[i] = TN_list[i] + 1
            elif predict_list[i][j] != Y_folds[i][j] and Y_folds[i][j] == 1:
                FN_list[i] = FN_list[i] + 1
            else:
                FP_list[i] = FP_list[i] + 1

    TP_prob = [(TP_list[0]) / ((TP_list[0]) + (FP_list[0])), (TP_list[1]) / ((TP_list[1]) + (FP_list[1])), (TP_list[2]) / ((TP_list[2]) + (FP_list[2])), (TP_list[3]) / ((TP_list[3]) + (FP_list[3])), (TP_list[4]) / ((TP_list[4]) + (FP_list[4]))]
    TN_prob = [(TN_list[0]) / ((TN_list[0]) + (FN_list[0])), (TN_list[1]) / ((TN_list[1]) + (FN_list[1])), (TN_list[2]) / ((TN_list[2]) + (FN_list[2])), (TN_list[3]) / ((TN_list[3]) + (FN_list[3])), (TN_list[4]) / ((TN_list[4]) + (FN_list[4]))]

    TPmean_prob = 0
    TNmean_prob = 0
    for i in range(0,5):
        TPmean_prob = TPmean_prob + float(TP_prob[i])
        TNmean_prob = TNmean_prob + float(TN_prob[i])
    TPmean_prob = TPmean_prob / 5
    TNmean_prob = TNmean_prob /5

    print("The mean probability of a positive being true over all splits for model", msl - 30, "is", TPmean_prob)
    print("The mean probability of a negative being true over all splits for model", msl - 30, "is", TNmean_prob)

    model_list = np.append(model_list, TPmean_prob) #TPmean_prob matters much more than TNmean_prob
    
#find the best model
best_model = 0
for i in range(0,20):
    if model_list[i] > model_list[best_model]:
        best_model = 0
        best_model = best_model + i
    else:
        pass

#Test the best model
dtc = DecisionTreeClassifier(min_samples_leaf = best_model + 30)
dtc.fit(X_train, Y_train)
final_test = dtc.predict(X_test)

TP_final = 0
TN_final = 0
FN_final = 0
FP_final = 0
for i in range(0,len(final_test)):
    if final_test[i] == Y_test[i] and Y_test[i] == 1:
            TP_final = TP_final + 1
    elif final_test[i] == Y_test[i] and Y_test[i] == 0:
            TN_final = TN_final + 1
    elif final_test[i] != Y_test[i] and Y_test[i] == 1:
            FN_final = FN_final + 1
    else:
            FP_final = FP_final + 1

TP_prob_final = TP_final / (FP_final + TP_final)
TN_prob_final = TN_final / (TN_final + FN_final)
print("The probability of the final model's positive prediction being true on unseen data is", TP_prob_final)
print("The probability of the final model's negative prediction being true on unseen data is", TN_prob_final)
```
